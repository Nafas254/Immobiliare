```python
import numpy as np
import pandas as pd 
import plotly as plt
import streamlit as st
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import plotly.express as px
import plotly.graph_objs as go
```

```python
df=pd.read_csv("https://frenzy86.s3.eu-west-2.amazonaws.com/python/data/immobili.csv")
```

```python
index_to_remove = 506
df=df.drop(index_to_remove)
```

```python
df=df.astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
sns.pairplot(df);
```

```python
y=df['medv']
X=df.drop(columns='medv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=800)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)

```

```python
model.coef_
model.intercept_
```

```python
lenght=y_pred.shape[0]
x=np.linspace(0,lenght,lenght)
```

```python
fig=plt.figure()

plt.plot(x, y_pred)
plt.plot(x, y_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
```

```python
print('R2_score: ', r2score)
print('MAE: ', mae)
print('MSE: ', mse)
print('RMSE: ', rmse)
 
```
